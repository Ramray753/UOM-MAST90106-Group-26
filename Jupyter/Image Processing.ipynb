## Thresholding 

```python
import cv2  
from matplotlib import pyplot as plt

# img_list = []
# for i in image_path:

img = cv2.imread('10003.jpg')  
 
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)  
ret, binary = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)  
 
img1, contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)  
# cv2.drawContours(img,contours,-1,(0,0,255),3)  



plt.imshow(img)
plt.show()

plt.imshow(img1,cmap = 'gray', interpolation = 'bicubic' )
plt.show()

cv2.imwrite("10003_n.jpg", img1)

# cv2.imshow("img", img)  
# cv2.waitKey(0)
```

```python
import os
import cv2
img_objects = {}
for i in os.listdir("train/cracked/"):
    if ".jpg" in i:
        #print(i)
        img = cv2.imread("train/cracked/"+str(i))
        #print(img)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        ret, binary = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)  
        img1, _contours, _hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)  
        img_objects[i] = img1 
```

```python
for i in img_objects.keys():    # save the processed results
    #plt.imshow(i, cmap="gray")
    #plt.show()
    print("train/new/new_{}".format(i))
    cv2.imwrite("train/new/new_{}".format(i), img_objects[i])
```

## Background Subtraction with KNN

```python
import os
import cv2
img_objects = {}
fgbg1 = cv2.createBackgroundSubtractorKNN()

for i in os.listdir("images/"):
    if ".JPG" in i:
        #print(i)
        img = cv2.imread("images/"+str(i))
        #print(img)
        fgmask1 = fgbg1.apply(img);
        img_objects[i] = fgmask1 
```

```python
for i in img_objects.keys():    # save the processed results
    #plt.imshow(i, cmap="gray")
    #plt.show()
    print("bg_sub_KNN_images/new_{}".format(i))
    cv2.imwrite("bg_sub_KNN_images/new_{}".format(i), img_objects[i])
```
