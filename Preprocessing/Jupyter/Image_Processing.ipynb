## Otsu’s thresholding algorithm

```python
# import the necessary packages
import argparse
import cv2
from matplotlib import pyplot as plt

# load the image and display it
image = cv2.imread("LCMS-Range-Shuffled/train/00405.JPG")
cv2.imshow("Image", image)
# convert the image to grayscale and blur it slightly
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (7, 7), 0)

# apply Otsu's automatic thresholding which automatically determines
# the best threshold value
(T, threshInv) = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)
cv2.imshow("Threshold", threshInv)
plt.show()
print("[INFO] otsu's thresholding value: {}".format(T))

# visualize only the masked regions in the image
masked = cv2.bitwise_and(image, image, mask=threshInv)
cv2.imshow("Output", masked)
cv2.waitKey(0)
```

```python
import os
import argparse
import cv2
from matplotlib import pyplot as plt

img_objects = {}
Threshold = {}
for i in os.listdir("LCMS-Range-Shuffled/train/"):
    if ".JPG" in i:
        #print(i)
        img = cv2.imread("LCMS-Range-Shuffled/train/"+str(i))
        # convert the image to grayscale and blur it slightly
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        blurred = cv2.GaussianBlur(gray, (7, 7), 0)
        # apply Otsu's automatic thresholding which automatically determines
        # the best threshold value
        (T, threshInv) = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)
        img_objects[i] = threshInv 
        Threshold[i] = T
```

```python
for i in img_objects.keys():    # save the processed results
    #plt.imshow(i, cmap="gray")
    #plt.show()
    print("LCMS-Range-Shuffled/Preprocess_train/OtsuThresholding/Otsu_{}".format(i))
    cv2.imwrite("LCMS-Range-Shuffled/Preprocess_train/OtsuThresholding/Otsu_{}".format(i), img_objects[i])
    print("[INFO] otsu's thresholding value: {}".format(Threshold[i]))
    
```

## adaptive thresholding

```python
# import the necessary packages
import argparse
import cv2
from matplotlib import pyplot as plt

# load the image and display it
image = cv2.imread("LCMS-Range-Shuffled/train/00405.JPG")
cv2.imshow("Image", image)
# convert the image to grayscale and blur it slightly
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (7, 7), 0)

# apply simple thresholding with a hardcoded threshold value
(T, threshInv) = cv2.threshold(blurred, 127, 255, cv2.THRESH_BINARY_INV)
cv2.imshow("Simple Thresholding", threshInv)
cv2.imwrite("LCMS-Range-Shuffled/VarThresh/Sim_00405.JPG",threshInv)
# import time
# time.sleep(10)
```

```python
# apply Otsu's automatic thresholding which automatically determines
# the best threshold value
(T, threshInv) = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)
cv2.imwrite("LCMS-Range-Shuffled/VarThresh/Otsu_00405.JPG",threshInv)
# cv2.imshow("Otsu Threshold", threshInv)
# plt.show()
# print("[INFO] otsu's thresholding value: {}".format(T))
```

```python
# instead of manually specifying the threshold value, we can use adaptive thresholding to 
# examine neighborhoods of pixels and adaptively threshold each neighborhood
thresh = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 21, 10)
cv2.imwrite("LCMS-Range-Shuffled/VarThresh/AdpM10_00405.JPG",thresh)
# cv2.imshow("Mean Adaptive Thresholding", thresh)
# plt.show()
```

```python
# perform adaptive thresholding again, this time using a Gaussian
# weighting versus a simple mean to compute our local threshold value
thresh2 = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 21, 4)
cv2.imwrite("LCMS-Range-Shuffled/VarThresh/check.JPG",thresh2)
# cv2.imshow("Gaussian Adaptive Thresholding", thresh)
```

## Pixel Propotion

```python
x,y= thresh2.shape

bk = 0
wt = 0

for i in range(x):
    for j in range(y):
        if thresh2[i,j]==0:
            bk+=1
        else:
            wt+=1
rate1 = wt/(x*y)
rate2 = bk/(x*y)

print("Write Pixel Proportion:", round(rate1*100,2),'%')
print("Black Pixel Proportion:", round(rate2*100,2),'%')
```

```python
import csv
import os
import numpy as np
from PIL import Image

datasets = []

with open("Pixel Proportion_Split.csv","w") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Image","Write Pixel","Black Pixel"])
    for filename in os.listdir("Origin-Binary/AdpGauTh_split/"):
        if filename.endswith('.jpeg'):
            image_path = os.path.join("Origin-Binary/AdpGauTh_split/", filename)
            image = np.array(Image.open(image_path))
            xy = image.shape

            rate_wt = np.count_nonzero(image)/(410 * 410)   # 255 value
            rate_bk = 1-rate_wt   # 0 value

            datasets.append([str(filename)[:-5],rate_wt,rate_bk])
    
    writer.writerows(sorted(datasets))
```

```python
import os
import argparse
import cv2
from matplotlib import pyplot as plt

img_objects = {}
Threshold = {}
for i in os.listdir("Origin-Binary/train/full/"):
    if ".jpeg" in i:
        #print(i)
        img = cv2.imread("Origin-Binary/train/full/"+str(i))
        # convert the image to grayscale and blur it slightly
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        blurred = cv2.GaussianBlur(gray, (7, 7), 0)
        # instead of manually specifying the threshold value, we can use
        # adaptive thresholding to examine neighborhoods of pixels and
        # adaptively threshold each neighborhood
        thresh = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 21, 10)
        img_objects[i] = thresh
```

## Background Subtraction with KNN

```python
import os
import cv2
img_objects = {}
fgbg1 = cv2.createBackgroundSubtractorKNN()

for i in os.listdir("LCMS-Range-Shuffled/train/"):
    if ".JPG" in i:
        #print(i)
        img = cv2.imread("LCMS-Range-Shuffled/train/"+str(i))
        #print(img)
        fgmask1 = fgbg1.apply(img);
        img_objects[i] = fgmask1 
```

```python
for i in img_objects.keys():    # save the processed results
    #plt.imshow(i, cmap="gray")
    #plt.show()
    print("LCMS-Range-Shuffled/pre_KNN/KNN_{}".format(i))
    cv2.imwrite("LCMS-Range-Shuffled/pre_KNN/KNN_{}".format(i), img_objects[i])
```
